```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
y=np.random.randint(1,100,10)
x,y
```

```python
plt.plot(x,y, color='red',linestyle=':', linewidth=3)
plt.title('Line Plot')
plt.xlabel('temp')
plt.ylabel('entropy')
plt.show()
```

```python
y1=np.random.randint(1,200,10)
```

```python
plt.plot(x,y, color='g',linestyle=':', linewidth=3)
plt.plot(x,y1, color='yellow',linestyle='--', linewidth=2)
plt.title('Line Plot')
plt.xlabel('temp')
plt.ylabel('entropy')
plt.grid()
plt.show()
```

```python
plt.subplot(2,1,1) #plt.subplot(1,2,1)
plt.plot(x,y, color='blue',linestyle=':', linewidth=3)
plt.subplot(2,1,2) #plt.subplot(1,2,2)
plt.plot(x,y1, color='red',linestyle='--', linewidth=2)
plt.show()
```

### Bar Plot-

```python
Students={'bob':34,'ani':77,'deep':98}
names=list(Students.keys())
marks=list(Students.values())
```

```python
plt.bar(names,marks)
plt.title('Bar Plot')
plt.xlabel('names')
plt.ylabel('marks')
plt.show()
```

```python
plt.barh(names,marks, color='yellow') #barh(x,y)
plt.title('Bar Plot')
plt.xlabel('names')
plt.ylabel('marks')
plt.grid()
plt.show()
```

### Scatter Plot-

```python
plt.scatter(x,y, marker='*', c='red', s=50)
plt.title('Scatter Plot')
plt.xlabel('temp')
plt.ylabel('entropy')
plt.show()
```

```python
plt.scatter(x,y, color='g',marker='*', s=300)
plt.scatter(x,y1, color='yellow',marker='.', s=200)
plt.title('scatter Plot')
plt.xlabel('temp')
plt.ylabel('entropy')
plt.grid()
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(x,y, color='g',marker='*', s=300)
plt.subplot(1,2,2)
plt.scatter(x,y1, color='yellow',marker='.', s=200)
plt.show()
```

### histogram-

```python
plt.hist(y, color='green', bins=4)
plt.show()
```

### Box Plot and Violin Plot-

```python
data=list([x,y,y1])
plt.boxplot(data)
plt.show()
```

```python
plt.violinplot(data, showmedians=True)
plt.show()
```

### Pie Chart-

```python
plt.pie(marks,labels=names, autopct='%0.1f%%', colors=['red','yellow','green'])
plt.show()
```
