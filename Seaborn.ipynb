### Scatterplot-

```python
sns.scatterplot(x='sepal_length',y='sepal_width', data=dataaa, hue='species', style='species' )
plt.show()
```

```python
sns.scatterplot(x='sepal_length',y='sepal_width', data=dataaa, hue='petal_length')
plt.show()
```

### Distplot or Histogram-

```python
sns.distplot(dataaa['petal_length'], color='green')
plt.show()
```

```python
sns.distplot(dataaa['petal_length'], hist= False)
plt.show()
```

```python
sns.distplot(dataaa['petal_length'],bins=10, kde=False)
plt.show()
```

```python
sns.distplot(dataaa['petal_length'], color='green', vertical=True)
plt.show()
```

### Jointplot-

```python
sns.jointplot(x='sepal_length', y='petal_length', data= dataaa, color='red')
plt.show()
```

```python
sns.jointplot(x='sepal_length', y='petal_length', data= dataaa, color='olive', kind='reg')
plt.show()
```

### Boxplot-

```python
sns.boxplot(x='species', y='petal_length', data= dataaa)
plt.show()
```

### Pairplot-

```python
sns.pairplot(dataaa, hue='species')
plt.show()
```
